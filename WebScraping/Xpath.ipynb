```python
def get_one_html(url):
    try:
        useragent=ua.random
        kv = {'User-Agent':useragent }
        r = requests.get(url,headers=kv)
        print(useragent,r.status_code)
        if r.status_code==200:
            r.encoding = 'utf-8'
            return r.text
    except:
        return None
```

```python
url ='https://www.douban.com'
get_one_html(url)
```
